```python
remove_columns = ([
  #Colunas categoricas
  'ID_CLIENTE',
  'TIPO_FUNCIONARIO',
  'ESTADO_NASCIMENTO',
  'CIDADE_NASCIMENTO',
  'ESTADO_RESIDENCIAL',
  'CIDADE_RESIDENCIAL',
  'BAIRRO_RESIDENCIAL',
  'CODIGO_A',
  'FLAG_TELEFONE_MOVEL',
  'ESTADO_PROFISSIONAL',
  'CIDADE_PROFISSIONAL',
  'BAIRRO_PROFISSIONAL',
  'CODIGO_AREA_TELEFONE_PROFISSIONAL',
  'FLAG_REGISTRO_ACSP',
  'CEP_RESIDENCIAL_3',
  'CEP_PROFISSIONAL_3',

  #Colunas nullas
  'QUANT_CARTOES_ADICIONAIS',
  'NIVEL_EDUCACIONAL',
  'FLAG_DOCUMENTO_RESIDENCIAL',
  'FLAG_RG',
  'FLAG_CPF',
  'FLAG_COMPROVANTE_RENDA'

])

df_adjusted = (
  df_raw
  .withColumn(
    'SEXO',
    when(col('SEXO') == 'M', 1)
    .when(col('SEXO') == 'F', 0)
    .otherwise(-1000)
  )
  .withColumn(
    'FLAG_TELEFONE_RESIDENCIAL',
    when(col('FLAG_TELEFONE_RESIDENCIAL') == 'Y', 1)
    .otherwise(0)
  )
  .withColumn(
    'FLAG_TELEFONE_PROFISSIONAL',
    when(col('FLAG_TELEFONE_PROFISSIONAL') == 'Y', 1)
    .otherwise(0)
  )
  .withColumn(
    'EMPRESA',
    when(col('EMPRESA') == 'Y', 1)
    .otherwise(0)
  )
  .withColumn(
    'TIPO_ENVIO_APLICACAO',
    when(col('TIPO_ENVIO_APLICACAO') == 'Carga', 1)
    .when(col('TIPO_ENVIO_APLICACAO') == 'Web', 2)
    .otherwise(0)
  )
).drop(*remove_columns)

df_adjusted.show()

# Cast e mediana
df_adjusted = df_adjusted.withColumn("MESES_RESIDENCIA", col("MESES_RESIDENCIA").cast("int"))
MEDIAN_MESES_RESIDENCIA = df_adjusted.approxQuantile("MESES_RESIDENCIA", [0.5], 0.01)[0]
print(MEDIAN_MESES_RESIDENCIA)
```

```python
def substituir_nulls_por_valor(df, valor_substituto=-1000):
    """
    Substitui todos os valores nulos do DataFrame Spark por um valor especificado.

    Parâmetros:
    - df: DataFrame Spark
    - valor_substituto: valor para substituir os nulls (default = -1000)

    Retorna:
    - Novo DataFrame com nulls substituídos
    """
    for coluna in df.columns:
        df = df.withColumn(
            coluna,
            when(col(coluna).isNull(), lit(valor_substituto))
            .when(col(coluna) == 'NULL', lit(valor_substituto))
            .otherwise(col(coluna))
        )
    return df


df_filled = (
  substituir_nulls_por_valor(df_adjusted)
)

df_filled.show()
```

```python
median_meses_residencia = df_filled.agg(avg("MESES_RESIDENCIA")).collect()[0][0]

df_transform = (
  df_filled
  .withColumn(
    'MESES_RESIDENCIA',
    when(col('MESES_RESIDENCIA') == -1000, lit(MEDIAN_MESES_RESIDENCIA))
    .otherwise(col('MESES_RESIDENCIA'))
  )
)

df_transform.filter(col('SEXO') != -1000).show()
```

```python
# 1. Coleta os dados como uma lista de Row
dados = df_transform.collect()
colunas = df_transform.columns
df_pandas = pd.DataFrame([r.asDict() for r in dados], columns=colunas)

# 2. Separar X e y
X = df_pandas.drop('ROTULO_ALVO_MAU=1', axis=1)
y = df_pandas['ROTULO_ALVO_MAU=1']

# 3. Corrigir tipos
for col in X.columns:
    if X[col].dtype == "object":
        try:
            X[col] = pd.to_numeric(X[col])
        except:
            X[col] = X[col].astype("category")

# 4. Train test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=50)

# 5. Treinar modelo
model_filter = XGBClassifier(enable_categorical=True, eval_metric='logloss')
model_filter.fit(X_train, y_train)

# 6. Avaliar
y_pred = model_filter.predict(X_test)

from sklearn.metrics import classification_report, ConfusionMatrixDisplay, confusion_matrix
print(classification_report(y_test, y_pred))

cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(cm)
disp.plot()
```

```python
from xgboost import XGBClassifier
from sklearn.metrics import classification_report, ConfusionMatrixDisplay, accuracy_score, confusion_matrix

# 1. Copiar X e y
X_train_embedded = X_train.copy()
y_train_embedded = y_train.copy()
X_test_embedded = X_test.copy()
y_test_embedded = y_test.copy()

# 2. Treinar o primeiro modelo para capturar importâncias
xgb = XGBClassifier(enable_categorical=True, eval_metric='logloss')
xgb.fit(X_train_embedded, y_train_embedded)

# 3. Pegar a importância das features
feat_importance = xgb.feature_importances_ * 100

# 4. Montar dicionário de importância
importance_dict = {}
for i, col in enumerate(X_train_embedded.columns):
    importance_dict[col] = feat_importance[i]

# Mostrar importâncias
print("\nImportância das features:")
for feature, importance in importance_dict.items():
    print(f"{feature}: {importance:.4f}")

# 5. Deletar colunas com importância < 1%
for col, importance in importance_dict.items():
    if importance < 1:
        X_train_embedded.drop(col, axis=1, inplace=True)
        X_test_embedded.drop(col, axis=1, inplace=True)
        print(f"❌ Deletado: {col}")

# 6. Treinar novo modelo
xgb = XGBClassifier(enable_categorical=True, eval_metric='logloss')
xgb.fit(X_train_embedded, y_train_embedded)

# 7. Predizer e avaliar
y_pred_embedded = xgb.predict(X_test_embedded)

print("\n📋 Classification Report:")
print(classification_report(y_test_embedded, y_pred_embedded))

# 8. Matriz de confusão
cm = confusion_matrix(y_test_embedded, y_pred_embedded)
disp = ConfusionMatrixDisplay(cm)
disp.plot()

# 9. Acurácia final
accuracy_embedded = (cm[0][0] + cm[1][1]) / cm.sum() * 100
print(f'✅ Accuracy usando Embedded Feature Selection: {accuracy_embedded:.2f}%')
```
